```python
#step3: Handling the missing data or Nan values

#Scenerio1: missing value in numerical column calculated by mean or average of the column
#Scenerio2: Missing values in non numeric values calculated by filling the empty value with highly repeated value
#Scenerio3: having the missing values in huge dataset(20000) Eg:2 rows having empty values in 20000 rows sol: Remove the empty values rows from dataset
import numpy as np
from sklearn.impute import SimpleImputer# SimpleImputer is a class which was imported from sklearn.impute
imputer=SimpleImputer(missing_values=np.nan,strategy='mean')
imputer.fit(X[:,1:3])
X[:,1:3]=imputer.transform(X[:,1:3])
```

```python
X
```

```python
# Step4: Handling the catogorial data
# Which converts the text data into numeric data
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
#LabelEncoder is a class which was imported from sklearn.preprocessing library
#It is used to convert categorial data to numeric data
# ColumnTransformer is a class imposed from sklearn.compose.
# It is used to apply changes to columns of an array.
#It replaces numeric data in the place of categorial data.Eg:Harika->1,0,0,0
ct=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[0])],remainder='passthrough')
# encoder is a method which coverts text to binary

X=ct.fit_transform(X)

```

```python
X
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
Y=le.fit_transform(Y)
```

```python
Y
```

```python
#step5:split the data into training and testing data
from sklearn.model_selection import train_test_split #It is a class, imported from sk.model_selection.
#It was used for splitting the data into training and testing
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
```

```python
#step6:feature scaling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train[:,3:]=sc.fit_transform(X_train[:,3:])
X_test[:,3:]=sc.transform(X_test[:,3:])
```

```python
#machine learning
# 1.supervisioned learning - when we have labelled data,
# and the model tasks are classification or prediction.
# 2.unsupervisioned learning - unlabelled data, model remembers the patterns.
# we can perform predictions also.
# 3.reinforcement learning - model will train by using trail and error method
# algorithms in supervision learning-1.linear regression,2.logistic regression
#logistic regression algorithm-when the model performs classification tasks
#linear regression - when model performs prediction tasks, we have to apply this algorithm
#step7: Applying algorithm to the data for model training.
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression()
classifier.fit(X_train,Y_train)
```

```python
Y_pred=classifier.predict(X_test)
```
